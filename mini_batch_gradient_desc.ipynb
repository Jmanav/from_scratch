```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import numpy as np
import random
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
class mb_GDregressor:
  def __init__(self,learning_rate,epochs,batch_size):
    self.intercept_ = None
    self.coef_ = None
    self.lr = learning_rate
    self.epochs = epochs
    self.batch_size = batch_size

  def fit(self,X_train,y_train):
    self.coef_ = np.ones(X_train.shape[1])
    self.intercept_ = 0

    for i in range(self.epochs):
      for j in range(int(X_train.shape[0]/self.batch_size)):
        #we'll be creating a batch
        idx = random.sample(range(X_train.shape[0]),self.batch_size)
        y_hat = np.dot(X_train[idx],self.coef_) + self.intercept_ #here X_train[idx] ->this is called fancy indexing

        intercept_der = -2 * np.mean(y_train[idx] - y_hat)
        self.intercept_ = self.intercept_ - (self.lr*intercept_der)

        coef_der = -2*np.dot((y_train[idx]-y_hat),X_train[idx])
        self.coef_ = self.coef_ - (self.lr * coef_der)
    print(self.coef_,self.intercept_)

  def predict(self,X_test):
    return np.dot(X_test,self.coef_) + self.intercept_
```

```python
idx = random.sample(range(X_train.shape[0]),60)
X_train[idx].shape
```

```python
reg = mb_GDregressor(0.1,1000,50)
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
