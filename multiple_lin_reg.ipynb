```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_diabetes
X,y = load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
class multiple_linear_regression():
  def _init_(self):
    #whats the diff betweeen self.coeff_ and self.coeff..what diff does the underscore make
    self.coeff=None
    self.intercept=None
  def fit(self,X_train,y_train):
    X_train = np.insert(X_train,0,1,axis=1)
    betas = np.linalg.inv(np.dot(X_train.T,X_train)).dot(X_train.T).dot(y_train)
    #w.k.t beta[0] is the itercept and the rest are the coeff
    self.intercept = betas[0]
    self.coeff = betas[1:]

  def predict(self,X_test):
    return self.intercept + np.dot(X_test,self.coeff)
```

```python
lr = multiple_linear_regression()
lr.fit(X_train,y_train)
```

we can see that we got the coefficients

```python
lr.predict(X_test)
```
